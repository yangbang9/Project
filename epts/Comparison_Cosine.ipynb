```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data/processed_data.csv', index_col = 0)
df
```

```python
## 사용자 정보 입력
player = {
    'Name' : 'Yang',
    'Position' : 'FW',
    'Acceleration' : 70,
    'SprintSpeed' : 80,
    'Agility' : 50,
    'Stamina' : 90,
    'Aggression' : 80

}
features = ['Acceleration', 'SprintSpeed', 'Agility', 'Stamina', 'Aggression']
# df.loc[0] = player

# df[features] = df[features].astype(float)
player = pd.DataFrame(player, index=[-1])
player
```

```python
df = pd.concat([player, df])
df['Total'] = df[features].sum(axis = 1)
df = df.reset_index(drop=True)
df
```

```python
type(df.iloc[1]['Name'])
```

```python
# 특성 값만 추출하여 벡터로 변환
features = ['Acceleration', 'SprintSpeed', 'Agility', 'Stamina', 'Aggression']
vectors = df[features].values

# 코사인 유사도 계산 함수
def cosine_similarity(vector1, vector2):
    dot_product = np.dot(vector1, vector2)
    norm_vector1 = np.linalg.norm(vector1)
    norm_vector2 = np.linalg.norm(vector2)
    similarity = dot_product / (norm_vector1 * norm_vector2)
    return similarity

# 비교할 선수 이름
target_player = 'Yang'
target_vector = df[df['Name'] == target_player][features].values[0]

# 다른 선수들과의 코사인 유사도 계산
similarities = []
for vector in vectors:
    similarity = cosine_similarity(target_vector, vector)
    similarities.append(similarity)

# 결과를 데이터프레임에 추가하여 유사도 순으로 정렬
df['cosine_similarity'] = similarities
sorted_df = df.sort_values(by='cosine_similarity', ascending=False)

sorted_df
```

```python
# 코사인 유사도 값을 [0, 100] 범위로 변환하는 함수
def convert_similarity_to_percentage(similarity):
    percentage = (similarity + 1) / 2 * 100
    return percentage

# 비교할 선수 이름
target_player1 = 'yang'
target_player2 = 'Lee Bum Soo'
vector1 = df[df['Name'] == target_player1][features].values[0]
vector2 = df[df['Name'] == target_player2][features].values[0]



# 두 선수 간의 코사인 유사도 계산
cosine_similarity_value = cosine_similarity(vector1, vector2)

# 코사인 유사도 값을 백분율로 변환
similarity_percentage = convert_similarity_to_percentage(cosine_similarity_value)

print(f"The similarity between the two players is: {similarity_percentage:.2f}%")
```
