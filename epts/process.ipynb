```python
import pandas as pd
import numpy as np
```

```python

file = pd.read_csv('data/FIFA23_official_data.csv', encoding_errors='ignore')
file.info()
```

```python
cols = ['Name','Position','Acceleration','SprintSpeed','Agility','Stamina','Aggression']
```

```python
file = file[cols]
temp = file.columns[2:7]
file['Total'] = file[temp].sum(axis=1)
```

```python
file = file.dropna(axis=0)
file.info()
```

```python
file.to_csv("./data/processed_data.csv")
```
