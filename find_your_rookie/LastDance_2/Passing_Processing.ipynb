```python
import pandas as pd
import numpy as np

df = pd.read_csv("passing.csv")
df.info()
```

```python
df1 = df[["name","age","position1","position2", "Apps","Mins", "Assists","KeyP","AvgP", "PS%", "Crosses",
          "LongB", "ThrB","Rating" ]]
```

```python
df1.head()
```

```python
print(df1.columns)
```

```python
## 25세 이하 선수들 df2

idx_age = df1[df1['age'] > 25].index
df2 = df1.drop(idx_age)
```

```python
## 출전 경기, 선발과 교체로 분할

df_app = df2[['Apps']]
df_app['start'] = df_app.Apps.str.split('(').str[0]
df_app['sub'] = df_app.Apps.str.split('(').str[1];
df_app['sub'] = df_app['sub'].replace('[\),]','', regex=True)
df_app = df_app.fillna(0)
df_app
```

```python
## 데이터 타입 형변환 (object > 정수형)
df_app['start'] = df_app['start'].astype(int)
df_app['sub'] = df_app['sub'].astype(int)
df_app.dtypes
```

```python
## 25세 이하 / 출전경기 19경기 이상인 선수
df_app2 = pd.DataFrame(df_app.sum(axis=1))
df_app2.columns = ["Apps"]

idx_app = df_app2[df_app2['Apps'] < 19].index
df_app3 = df_app2.drop(idx_app)
df_app3
```

```python
df4 = df1.loc[df_app3.index]
df4 = df4.fillna('')
df4
```

```python
#포지션별 분류
#포지션이 미드필더 중 나이가 25세 이하, 출전수가 19경기 이상인 선수
dataFilter1 = df4['position1'].str.contains("AM|DMC|M")
df_a = df4[dataFilter1]
dataFilter4 = df_a['position2'].str.contains("A|F")
df5 = df_a[~dataFilter4]
df6 = pd.DataFrame(df5)
df6.to_csv("passing_processing.csv")
df6
```
